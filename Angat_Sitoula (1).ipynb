```python
#Selecting important variable based on Analysis
df_final=df[['instant', 'season', 'yr','weathersit', 'temp', 'atemp','casual', 'registered', 'cnt']]
```

```python
X=df_final.drop("cnt",axis=1) #Creating Independent variable
y=df_final["cnt"] #Creatig Independent Variable
```

# **Train Test Split**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

# **Linear Regression Model Building**

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(X_train, y_train)
print(reg.score(X_test, y_test))
y_pred=reg.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```
